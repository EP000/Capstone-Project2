```python
#arr = np.array(Image.open("./data_cap/Original_data/train/0006R0_f01530.png"))
def sum_pixel(image_data):
    
    for c, i in palette.items():
        label = label_list[i]
        m = np.all(image_data == np.array(c).reshape(1,1,3), axis =2)
        label_sum.loc[label_sum['label']==label, 'sum']+=sum(sum(m))    
```
