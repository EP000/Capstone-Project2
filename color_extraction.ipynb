```python
from PIL import Image 
import cv2

im = Image.open('./data/Labeled_data/0001TP_006690.png')
rgb_im = im.convert('RGB')

img = cv2.imread('./data/Labeled_data/0001TP_006690.png')
img_row = img.shape[0] # 720
img_col = img.shape[1] # 960

img_pixel = []
for r in range(img_row):
    for c in range(img_col):
        img_pixel.append(rgb_im.getpixel((c, r)))
```

```python
img_pixel = list(set([tuple(pixel) for pixel in img_pixel]))
```

```python
import numpy as np
color_set = [np.array([[img]*25]*25) for img in img_pixel]
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3, 5, figsize = (16, 8), constrained_layout = True)

for i in range(3):
    for j in range(5):
        axs[i][j].imshow(color_set[(i*5)+j])
        axs[i][j].grid(False)
```
