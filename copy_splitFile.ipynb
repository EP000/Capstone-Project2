```python
import numpy as np
import pandas as pd
import os
import shutil
```

```python
train = []
f = open("./dataset/train.txt", 'r')
lines = f.readlines()
for line in lines:
    train.append(line.split('\n')[0]+'.png')
f.close()

val = []
f = open("./dataset/val.txt", 'r')
lines = f.readlines()
for line in lines:
    val.append(line.split('\n')[0]+'.png')
f.close()

test = []
f = open("./dataset/test.txt", 'r')
lines = f.readlines()
for line in lines:
    test.append(line.split('\n')[0]+'.png')
f.close()
```

```python
for move_f in train:
    shutil.copy('./data_cap/Original_data/'+move_f, './data_cap/Original_data/train/'+move_f)
    shutil.copy('./data_cap/Gray_data/'+move_f, './data_cap/Gray_data/train/'+move_f)
    
for move_f in val:
    shutil.copy('./data_cap/Original_data/'+move_f, './data_cap/Original_data/val/'+move_f)
    shutil.copy('./data_cap/Gray_data/'+move_f, './data_cap/Gray_data/val/'+move_f)

for move_f in test:
    shutil.copy('./data_cap/Original_data/'+move_f, './data_cap/Original_data/test/'+move_f)
    shutil.copy('./data_cap/Gray_data/'+move_f, './data_cap/Gray_data/test/'+move_f)
```
